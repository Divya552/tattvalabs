```python
import matplotlib 
from matplotlib import pyplot as pl,style
import pandas as pd
style.use('grayscale')
d=pd.read_csv('points1.csv')
n=len(d)
for i in range(0,n):
    pl.plot(d['x'],d['y'],'b')
pl.xlabel('X-axis')
pl.ylabel('Y-axis')
pl.title(' LINE GRAPH')
pl.show()
```

```python
import matplotlib 
from matplotlib import pyplot as pl,style
import pandas as pd
style.use('ggplot')
d=pd.read_csv('points1.csv')
n=len(d)
for i in range(0,n):
    pl.scatter(d['x'],d['y'],color='r')
pl.xlabel('X-axis')
pl.ylabel('Y-axis')
pl.title(' SCATTER GRAPH')
pl.show()
```

```python
import matplotlib 
from matplotlib import pyplot as pl,style
import pandas as pd
style.use('seaborn-notebook')
d=pd.read_csv('points1.csv')
n=len(d)
for i in range(0,n):
    pl.bar(d['x'],d['y'],color='c')
pl.xlabel('X-axis')
pl.ylabel('Y-axis')
pl.title('BAR GRAPH')
pl.show()
```

```python
import pandas as pd
import matplotlib 
from matplotlib import pyplot as pl,style
from mpl_toolkits.mplot3d import Axes3D
style.use('dark_background')
fig=pl.figure()
ax=fig.add_subplot(111,projection='3d')
x=[2,3,7,8,9]
y=[4,6,8,4,3]
z=[3,7,6,9,2]
ax.p(x,y,z,c='r',marker='o')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
pl.show()
```

```python
import matplotlib
from matplotlib import pyplot as pl,style
import numpy as np
style.use('ggplot')
x=np.arange(0,10)
y=np.exp(-x/10 ) * np.sin(180/np.pi*x)
z=x*np.exp(-x/3)
pl.plot(y,z)
pl.show()
```

```python
import matplotlib
from matplotlib import pyplot as pl,style
from mpl_toolkits.mplot3d import Axes3D
import csv
import pandas as pd
style.use('grayscale')
fig=pl.figure()
ax=fig.add_subplot(111,projection='3d')
d=pd.read_csv(r'C:\Users\hp\Downloads\iris.csv')
n=len(d)
for i in range(0,n):
    if(str(d.iloc[i]['species'])=='setosa'):
        ax.scatter(d.iloc[i]['sepal_length'],d.iloc[i]['sepal_width'],color='r')
    if(str(d.iloc[i]['species'])=='versicolor'):
        ax.scatter(d.iloc[i]['sepal_length'],d.iloc[i]['sepal_width'],color='b')
    if(str(d.iloc[i]['species'])=='virginica'):
        ax.scatter(d.iloc[i]['sepal_length'],d.iloc[i]['sepal_width'],color='y')
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
pl.savefig('C:\\Users\\hp\\Desktop\\iris_scatter_pic.jpg')
pl.show()



```
