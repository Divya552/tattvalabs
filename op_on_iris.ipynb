```python
import pandas as pd
import sqlite3
conn=sqlite3.connect('my__iris.sqlite')
cursor=conn.cursor()
df = pd.read_csv('Downloads\iris.csv')
df.to_sql('iris', conn, if_exists='append', index=False)
```

```python
for row in cursor.execute("SELECT * FROM iris WHERE species='setosa'"):
    print('sepal_length',row[0])
    print('sepal_width',row[1])
    print('petal_length',row[2])
    print('petal_width',row[3])
    
```

```python
cursor.execute("UPDATE iris SET sepal_length=10.0 WHERE sepal_width=3.8 AND petal_length=1.5 AND species='setosa'")
conn.commit()
```

```python
cursor.execute("DELETE FROM iris WHERE species='versicolor'")
conn.commit()
```
